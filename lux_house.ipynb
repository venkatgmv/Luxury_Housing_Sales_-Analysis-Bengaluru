```python
df.drop_duplicates(inplace=True)
```

```python
df['Ticket_Price_Cr']=df['Ticket_Price_Cr'].str.replace('Cr','',regex=False).str.strip().str.replace('Rs','').str.replace('â‚¹','')
#remove 'Cr','Rs','â‚¹' and extra spaces from the column
df['Ticket_Price_Cr']=pd.to_numeric(df['Ticket_Price_Cr'],errors='coerce')
#convert the column to numeric, setting errors to NaN
```

```python
df['Ticket_Price_Cr']
```

```python
df['Unit_Size_Sqft'] = df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].mean())
# fill values missed in 'Unit_Size_Sqft' with the mean of the column
df['Ticket_Price_Cr'] = df['Ticket_Price_Cr'].fillna(df['Ticket_Price_Cr'].mean())
# Fill values missed in 'Ticket_Price_Cr' with the mean of the column
df['Amenity_Score'] = df['Amenity_Score'].fillna(df['Amenity_Score'].median())
# Fill values missed in 'Amenity_Score' with the median of the column
df['Buyer_Comments'] = df['Buyer_Comments'].fillna("No Comments")
# fill values missed in 'Buyer_Comments' with "No Comments"
```

```python
df
```

```python
df.dtypes
```

```python
norm_column = ['Micro_Market', 'Developer_Name', 'Configuration',
             'Transaction_Type', 'Buyer_Type', 'Possession_Status',
             'Sales_Channel', 'NRI_Buyer']
# Columns to be normalized using one-hot encoding
```

```python
for col in norm_column:
    df[col]=df[col].str.strip().str.title()
# Standardize text data by stripping extra spaces and converting to title case
```

```python
df["Price_per_Sqft"] = df["Ticket_Price_Cr"] * 10000000 / df["Unit_Size_Sqft"]
# Create a new column 'Price_per_Sqft' by calculating price per square foot
```

```python
# Convert the 'Purchase_Quarter' column to datetime format
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
# Extract the quarter from the 'Purchase_Quarter' column and create a new column 'Quarter_Number'
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
# Display the updated DataFrame with the new 'Quarter_Number' column
df['Quarter_Number']
```
